```python
def buscaSequencial(x, v):
  n = len(x)
  for i in range(0, n):
    if v == x[i]:
      return i
  return -1
```

```python
def buscaBinaria(x, v):
  li = 0
  ls = len(x) - 1
  while li <= ls:
    i = (li + ls) // 2
    if v == x[i]:
      return i
    elif v < x[i]:
      ls = i - 1
    else:
      li = i + 1
  return -1
```

```python
p = 6
#ret = buscaSequencial([2, 4, 7, 1, 3, 8, 9, 10, 0], p)
ret = buscaBinaria([2, 5, 6, 7, 9, 12, 15], p)
if ret != -1:
  print(p, "está na posição", ret)
else:
  print("Valor não encontrado")
```
