```python
def qa(x):
  n = len(x)
  for i in range(0, n-1):
    for j in range(i + 1, n):
      if x[j] < x[i]:
        x[i], x[j] = x[j], x[i]
    print(x)
  return x
```

```python
x = [5, 2, 4, 3, 1]
x = qa(x)
```

```python
def selectionSortO(x):
  n = len(x)
  for i in range(0, n-1):
    m = i
    for j in range(i + 1, n):
      if x[j] < x[m]:
        m = j
    x[i], x[m] = x[m], x[i]
  return x
```

```python
a = [1, 2, 3, 4, 5]
a, k = selectionSortO(a)
print("Executou o laço mais interno", k," vezes")
```
