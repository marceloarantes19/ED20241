```python
def qa(x):
  n = len(x)
  for i in range(0, n-1):
    for j in range(n-1, i, -1):
      if x[j] < x[j-1]:
        x[j], x[j-1] = x[j-1], x[j]
    print(x)
  return x
```

```python
x = [5, 2, 4, 3, 1]
x = qa(x)
```
