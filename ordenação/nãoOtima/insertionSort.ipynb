```python
def qa(x):
  n = len(x)
  for i in range(1, n):
    v = x[i]
    j = i
    while j > 0 and v < x[j-1]:
      x[j] = x[j - 1]
      j = j - 1
    x[j] = v
    print(x)
  return x
```

```python
x = [5, 2, 4, 3, 1]
x = qa(x)
```
