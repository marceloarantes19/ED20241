```python
def mHeap(x, atu, fHeap):
  esq = atu * 2 + 1
  dir = atu * 2 + 2
  ma  = atu
  if dir < fHeap and x[dir] > x[ma]:
    ma = dir
  if esq < fHeap and x[esq] > x[ma]:
    ma = esq
  if ma != atu:
    x[atu], x[ma] = x[ma], x[atu]
    mHeap(x, ma, fHeap)
```

```python
def heapSort(x):
  for i in range(len(x)//2,-1,-1):
    mHeap(x, i, len(x))
  for i in range(len(x)-1, 0, -1):
    x[0], x[i] = x[i], x[0]
    mHeap(x, 0, i)
```

```python
x = [8, 10, 15, 7, 5, 25, 3, 20]
#x = [8, 10, 15, 7, 25, 3, 20]
heapSort(x)
print(x)
```

```python
def mergeSort(x):
  merge(x, 0, len(x))
```

```python
def merge(x, inicio, fim):
  if (fim - inicio) > 1:
    meio = (inicio + fim) // 2
    merge(x, inicio, meio)
    merge(x, meio, fim)
    intercala(x, inicio, meio, fim)
```

```python
def intercala(x, inicio, meio, fim):
  vetA = x[inicio:meio]
  vetB = x[meio:fim]
  indA = 0
  indB = 0
  for i in range(inicio, fim):
    if (indB >= len(vetB)) or (indA < len(vetA) and vetA[indA] < vetB[indB]):
      x[i] = vetA[indA]
      indA = indA + 1
    else:
      x[i] = vetB[indB]
      indB = indB + 1
```

```python
x = [0, 8,7,6,5,4,6, 7, 9, 3,2,1]
mergeSort(x)
print(x)
```
